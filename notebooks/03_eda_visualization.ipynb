# 03 - Phân tích trực quan
Notebook này đọc `master.parquet` để tạo tối thiểu ba biểu đồ đáp ứng rubric: phân bố rating, tần suất thể loại, top phim và heatmap user-movie. Các biểu đồ được lưu trong `reports/figures`.

```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path('..').resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
from src.visualization import (
    plot_rating_distribution,
    plot_top_genres,
    plot_top_movies_by_ratings,
    plot_rating_heatmap,
 )

PROCESSED_DIR = PROJECT_ROOT / 'data' / 'processed'
FIG_DIR = PROJECT_ROOT / 'reports' / 'figures'
FIG_DIR.mkdir(parents=True, exist_ok=True)
master_df = pd.read_parquet(PROCESSED_DIR / 'master.parquet')
master_df[['userId', 'movieId', 'rating', 'genre_list']].head()
```

## Chuẩn bị dữ liệu & thư mục lưu hình
Đọc `master.parquet`, khởi tạo thư mục `reports/figures` và xem nhanh một vài cột quan trọng.

```python
rating_dist_path = FIG_DIR / 'rating_distribution.png'
plot_rating_distribution(master_df, output=rating_dist_path)
rating_dist_path
```

### Phân bố điểm số rating
Histogram + KDE giúp đánh giá xu hướng chung của người dùng khi chấm điểm.

```python
genre_path = FIG_DIR / 'genre_frequency.png'
plot_top_genres(master_df, top_n=10, output=genre_path)
genre_path
```

### Top thể loại được xem nhiều nhất
Đếm số lần xuất hiện từng thể loại (split từ cột `genre_list`) và hiển thị Top 10.

```python
top_movies_path = FIG_DIR / 'top_movies.png'
plot_top_movies_by_ratings(master_df, min_count=150, output=top_movies_path)
top_movies_path
```

### Top phim theo điểm trung bình (lọc theo lượt đánh giá)
Chỉ giữ các phim có ≥150 lượt chấm để tránh nhiễu rồi hiển thị Top 15 theo `avg_rating`.

```python
heatmap_path = FIG_DIR / 'rating_heatmap.png'
plot_rating_heatmap(master_df, output=heatmap_path)
heatmap_path
```
