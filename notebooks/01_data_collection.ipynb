# 01 - Thu thập dữ liệu
Notebook này dùng thư viện `src.data_ingestion` để đọc bộ MovieLens latest-small đã tải về trong `data/raw`, kiểm tra nhanh cấu trúc bảng và ghi lại file CSV chuẩn cho các bước kế tiếp.

```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path('..').resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
from src import data_ingestion

RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
PROCESSED_DIR = PROJECT_ROOT / 'data' / 'processed'
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
RAW_DIR
```

## Khởi tạo đường dẫn
Tạo biến trỏ đến thư mục dự án, thư mục raw và processed đồng thời đảm bảo đã tồn tại.

```python
raw_data = data_ingestion.load_raw_data(RAW_DIR)
movies_df = raw_data['movies']
ratings_df = raw_data['ratings']
movies_df.head()
```

## Nạp dữ liệu thô và xem nhanh
Đọc các bảng chính (movies, ratings) rồi hiển thị vài dòng đầu để xác nhận cấu trúc.

```python
ratings_df.head()
```

### Xem thêm vài bản ghi ratings
Bảng ratings thường lớn nên chỉ kiểm tra nhanh một lát cắt nhỏ để tránh tràn bộ nhớ.

```python
movies_path = PROCESSED_DIR / 'movies.csv'
ratings_path = PROCESSED_DIR / 'ratings.csv'
movies_df.to_csv(movies_path, index=False)
ratings_df.to_csv(ratings_path, index=False)
{"movies_path": str(movies_path), "ratings_path": str(ratings_path)}
```
